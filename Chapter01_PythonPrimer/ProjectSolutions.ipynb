```python
#P-1.32

result=0
cnt=0
while (True):
    n = int(input())
    operator = input()
    if n and operator:
        if operator=='+':
            cnt+=1
            result=result+n
        elif operator=='-':
            cnt+=1
            result=n-result
        elif operator=='*':
            cnt+=1
            if (cnt>1 and result!=0):
                result=result*n
            else:
                result=(result+1)*n
        elif operator=='/':
            cnt+=1
            if (cnt>1 and result!=0):
                result=result/n
            else:
                result=n/(result+1)
        elif operator=='=':
            print (result)
            break
    else:
        print ('Invalid entry in either of the two entities')
        break
```
